```python
class Plantation:
    
    def __init__(self,listofplants):
        self.availableplants=listofplants

    def displayAvailableplants(self):
        print("The plants we have in our plantation are as follows:")
        print("================================")
        for plant in self.availableplants:
            print(plant)

    def deleteplant(self,requestedplant):
        if requestedplant in self.availableplants:
            print("The plant you requested has now been removed.")
            self.availableplants.remove(requestedplant)
        else:
            print("Sorry the plant you have requested to delete, is currently not in the library.")

    def addplants(self,returnedplant):
        self.availableplants.append(returnedplant)
        print("New plant is Added Successfully.")


class Employee:

    def delplant(self):
        print("Enter the name of the plant you'd like to delete:")
        self.plant=input()
        return self.plant

    def newplant(self):
        print("Enter the name of the plant you'd like to add:")
        self.plant=input()
        return self.plant

def call():
    plantation=Plantation(["plant 1","plant 2","plant 3","plant 4","plant 5"]) 
    employee=Employee() 
    done=False 
    while done==False:
        print(""" ======Plants Menu=======
          1. Display all available plants
          2. Add a plants
          3. Delete a plants
          4. Exit
          """)
        try:
            choice=int(input("Enter Choice:"))
            if choice==1:
                plantation.displayAvailableplants()
            elif choice==2:
                plantation.addplants(employee.newplants())
            elif choice==3:
                plantation.deleteplants(employee.delBook())
            elif choice==4:
                break 
                            
        except:
            print("Please enter proper choice.")

call()
```

```python
class Library:
    
    def __init__(self,listofbooks):
        self.availablebooks=listofbooks

    def displayAvailablebooks(self):
        print("The books we have in our library are as follows:")
        print("================================")
        for book in self.availablebooks:
            print(book)

    def deleteBook(self,requestedBook):
        if requestedBook in self.availablebooks:
            print("The book you requested has now been removed.")
            self.availablebooks.remove(requestedBook)
        else:
            print("Sorry the book you have requested to delete, is currently not in the library.")

    def addBook(self,returnedBook):
        self.availablebooks.append(returnedBook)
        print("New Book is Added Successfully.")


class Librarian:

    def delBook(self):
        print("Enter the name of the book you'd like to delete:")
        self.book=input()
        return self.book

    def newBook(self):
        print("Enter the name of the book you'd like to add:")
        self.book=input()
        return self.book

def call():
    library=Library(["Book 1","Book 2","Book 3","Book 4","Book 5"]) 
    librarian=Librarian() 
    done=False 
    while done==False:
        print(""" ======LIBRARY MENU=======
          1. Display all available books
          2. Add a book
          3. Delete a book
          4. Exit
          """)
        try:
            choice=int(input("Enter Choice:"))
            if choice==1:
                library.displayAvailablebooks()
            elif choice==2:
                library.addBook(librarian.newBook())
            elif choice==3:
                library.deleteBook(librarian.delBook())
            elif choice==4:
                break 
                            
        except:
            print("Please enter proper choice.")

call()
```
