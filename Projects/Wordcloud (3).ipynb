```python
from wordcloud import WordCloud
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('wine.csv',index_col=0)
```

```python
df
```

```python
text=df['description']
```

```python
text
```

```python
text[2]
```

```python
text[1]
```

```python
t="welcome to java"
```

```python
wd=WordCloud(collocations=False).generate(text[1])
```

```python
wd
```

```python
plt.imshow(wd)
plt.axis("off")
plt.show()
```

```python
wd2=WordCloud().generate(text[1])
plt.imshow(wd2)
plt.axis("off")
plt.show()
```

```python
wd3=WordCloud(collocations=True).generate(text[1])
plt.imshow(wd3)
plt.axis("off")
plt.show()
```

```python
wd1=WordCloud(width=400,height=400,background_color='white').generate(text[2])
```

```python
plt.imshow(wd1,interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
plt.imshow(wd1,interpolation='bicubic')
plt.axis("off")
plt.show()
```

```python
plt.imshow(wd1,interpolation='quadric')
plt.axis("off")
plt.show()
```

```python
plt.imshow(wd1,interpolation='gaussian')
plt.axis("off")
plt.show()
```
